```python
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x = np.linspace(-10,10,500)
f = (2/gamma(0.25))*np.exp(-np.power(x,4))
plt.plot(x,f)
plt.grid()
plt.savefig('PDF.png')
plt.show()
```

```python
from scipy.special import gammaincc
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x_positive = np.linspace(0,10,250)
x_negative = np.linspace(-10,0,250)

cdf_positive = 1-0.5 *gammaincc(0.25,np.power(x_positive,4))
cdf_negative = 0.5*gammaincc(0.25,np.power(x_negative,4))
plt.plot(x_negative,cdf_negative,'b')
plt.plot(x_positive,cdf_positive,'b')
plt.grid()
plt.savefig('CDF.png')
plt.show()
```
