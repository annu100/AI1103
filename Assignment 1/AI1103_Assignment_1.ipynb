```python
from scipy.special import gamma, gammaincc, gammainccinv
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

x = np.linspace(-10,10,500)
f = (2/gamma(0.25))*np.exp(-np.power(x,4))
plt.plot(x,f)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('PDF.png')
plt.show()
```

```python
x_positive = np.linspace(0,10,250)
x_negative = np.linspace(-10,0,250)


c = 2/gamma(0.25)
cdf_positive = 1 - (c/4)*gamma(0.25)*gammaincc(0.25,np.power(x_positive,4))
cdf_negative = (c/4)*gamma(0.25)*gammaincc(0.25,np.power(x_negative,4))


plt.plot(x_negative,cdf_negative,'b')
plt.plot(x_positive,cdf_positive,'b')
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('CDF.png')
plt.show()
```

```python
c = 2/gamma(0.25)

y_firsthalf = np.linspace(0,0.5,250)
K = y_firsthalf*(4/c)*(1/gamma(0.25))
x_negative = -np.power(gammainccinv(0.25,K),0.25)



y_secondhalf = np.linspace(0.5,1,250)
K = (1-y_secondhalf)*(4/c)*(1/gamma(0.25))
x_positive = np.power(gammainccinv(0.25,K),0.25)
```

```python
X = np.concatenate((x_negative,x_positive))

x_list = np.linspace(-10,10,500)
P = []
for x in x_list:
    p = np.where(X<=x)[0].size / X.size
    P.append(p)
plt.plot(x_list,P,'.')
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
plt.savefig('CDF_Simulation.png')
plt.show()
```
